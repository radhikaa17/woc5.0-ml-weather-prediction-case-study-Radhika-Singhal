```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("/content/CarPrice_Assignment.csv")
data_car = data.drop('car_ID',axis = 1) #column of no use
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='carbody',hue='doornumber',data=data_car)
plt.xlabel('Car Body')
plt.ylabel('No.of cars')
plt.show()
```

This graph is showing that all the hardtop and convertible cars have two doors. 

All wagon cars have four doors.

Hatchback has mostly two doors and Sedan has mostly four.

```python
sns.lmplot( x="carlength", y="enginesize", data=data_car, fit_reg=False, legend=True,hue = 'fueltype')
```

As we can see above the positive relationship between Car length and Engine Size, as the car size increases, engine size is also increasing.

```python
plt.figure(figsize=(2,2))
sns.lmplot( x="price", y="carlength", data=data_car, fit_reg=False, legend=True)
sns.lmplot( x="price", y="carwidth", data=data_car, fit_reg=False, legend=True)
sns.lmplot( x="price", y="carheight", data=data_car, fit_reg=False, legend=True)
sns.lmplot( x="price", y="curbweight", data=data_car, fit_reg=False, legend=True)
sns.lmplot( x="price", y="enginesize", data=data_car, fit_reg=False, legend=True)
sns.lmplot( x="price", y="horsepower", data=data_car, fit_reg=False, legend=True)
```

Above graphs are showing relationship of car price with different attributes.

```python
plt.figure(figsize=(10,10))
corr = data_car.corr()
sns.heatmap(corr,annot=True)
```

Heatmap is showing degree of correlation between all numerical attributes. Darker colours are representing stronger positive(red) and negative(black) relationship while light colours representing weak negative relationships. 

The diagonal is 1 because it is showing relationship of a variable with itself which is always 1.

A boxplot gives us five number summary:

1- Minimum (Q0 or 0th percentile): the lowest data point in the data set excluding any outliers

2- Maximum (Q4 or 100th percentile): the highest data point in the data set excluding any outliers

3- Median (Q2 or 50th percentile): the middle value in the data set

4- First quartile (Q1 or 25th percentile): also known as the lower quartile qn(0.25), it is the median of the lower half of the dataset.

5- Third quartile (Q3 or 75th percentile): also known as the upper quartile qn(0.75), it is the median of the upper half of the dataset.[7]

In addition to above, boxplots also gives us outliers.

```python
fig = plt.figure(figsize =(5,5))
plt.boxplot(data_car['price'])
plt.show()
```

In price boxplot, cars priced more than 28,000 are outliers.
Median price of a car is 10,000.
