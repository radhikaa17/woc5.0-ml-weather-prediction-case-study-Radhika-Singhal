##2) How many values are missing for each column and try dropping or replacing NULL values using appropriate pandas’ function.

```python
# count of missing values in each column
data.isnull().sum()
```

Creator Name, Creator Gender, No of Likes, Language of the Video, Subtitle, Date of the Last Comment, No of Playlist have missing values.

Total 7 columns

```python
# total number of missing values in the dataframe or whole data
data.isnull().sum().sum()
```

```python
print(data.mean(numeric_only=True)) #finding mean of numeric columns
```

```python
# replacing na values in no of likes with its mean values
data["No of Likes"].fillna(584010.065265, inplace = True)
# replacing na values in no of playlist with mean no
data["No of Playlist"].fillna(82.246652, inplace = True)
```

```python
#finding mode for categorical columns
print(data['Creator Name'].mode())
print(data['Creator Gender'].mode())
print(data['Language of the Video'].mode())
print(data['Subtitle'].mode())
print(data['Date of the Last Comment'].mode())
```

```python
#filling using mode since all are categorical in nature
data["Creator Name"].fillna("Ganesh Prasad", inplace = True)
data["Creator Gender"].fillna("Male", inplace = True)
data["Language of the Video"].fillna("English", inplace = True)
data["Subtitle"].fillna("Yes", inplace = True)
data["Date of the Last Comment"].fillna("9/23/2022", inplace = True)
```

```python
data.isnull().sum().sum() #checking once again whether there are any missing value left
```

No missing values.

```python
#finding unique values in the column
#to find whether there are any null values other than NaN or any other abnormal value
for i in data.columns:
  print(data[i].value_counts())
  print()
```

'Premiered or Not' , 'Subtitle', 'Video Description' have some mistakes. Like No and NO are treated differently, hence we need to correct them.

```python
data['Subtitle'] = data['Subtitle'].replace(to_replace={'yes':'Yes'})
data['Video Description'] = data['Video Description'].replace(to_replace={'NO':'No'})
data['Premiered or Not'] = data['Premiered or Not'].replace(to_replace={'yes':'Yes'})
```

```python
#checking 
for i in data.columns:
  print(data[i].value_counts())
  print()
```
